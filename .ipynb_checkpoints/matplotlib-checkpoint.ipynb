```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(1, 5, 11) #membuat data array dimulai dari angka 1 sampai 5 dengan jumlah index ada 11
y = x ** 2 #dari array yang telah dibuat diatas  di beri pangkat 2
x
```

```python
y
```

```python
plt.plot(x, y, 'r') #r merupakan warna yaitu red atau merah
plt.xlabel('X Axis') #label untuk sumbu X
plt.ylabel('Y Axis') #label untuk sumbu Y
plt.title('Data Contoh') #Label untuk judul diagram
plt.show() #untuk memunculkan diagram
```

```python
#membuat banyak diagram atau plot di 1 canvas
plt.subplot(1,2,1) #1 adalah raw #2 adalah kolom #1 adlah plot
plt.plot(x, y, 'r--') #r-- adlah garis warna merah putus-putus
plt.subplot(1,2,2) #1 adalah raw #2 adalah kolom #2 adlah plot
plt.plot(y, x, 'b*-') #b*- adlah garis warna biru dengan campuran bintang dan garis
```

```python
fig = plt.figure() #membuat canvas kosong
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
#0.1 jarak dari kiri
#0.1 jarak dari bawah
#0.8 Lebar
#0.8 tinggi

axes.plot(x, y, 'b')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Contoh Judul')
```

```python
#PLOT DIDALAM PLOT
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #plot utama
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) #plot didalam

#pengaturan plot utama
axes1.plot(x,y, 'b')
axes1.set_xlabel('X utama')
axes1.set_ylabel('y utama')
axes1.set_title('Plot Utama')

#pengaturan plot dalam
axes2.plot(y,x, 'r')
axes2.set_xlabel('X dalam')
axes2.set_ylabel('y dalam')
axes2.set_title('Plot dalam');
```

```python
#figzise adalah pengaturan ukuran terhaddap canvas yang akan dibuat
#dpi adalh bentukan dari suatu ukuran hasil cetak atau visual semakin tinggi maka semakin besar gambarnya
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(8,4))

axes.plot(x, y, 'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('title')
```

```python
fig, axes = plt.subplots(figsize=(8,4))

axes.plot(x, y, 'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('title')

#save atau export figure menjadi format PNG, JPG, SVG, PGF, dan PDF.
fig.savefig("contoh.png", dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="Data x pangkat 2")
ax.plot(x, x**3, label="Data x pangkat 3")

#menambah info kedalam figure
ax.legend(loc=0)
#loc = posisi diman kita akan menaruhnya
# 0 default kiri atas
#1 = kanan atas
#2 = kiri atas
#3 = kiri bawah
#4 = kanan bawah
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x+1, color="red", linewidth=0.25) #color mengatur warna garis
ax.plot(x, x+2, color="red", linewidth=0.50) #linewidth mengatur ketebalan gsris
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, x+5, color="green", lw=3, linestyle='-') #bisa kita singkat
ax.plot(x, x+6, color="red", lw=3, ls='-.')  #lw = linewidth
ax.plot(x, x+7, color="yellow", lw=3, ls=':')  #ls = linestyle
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

#custum dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5,5,15,5]) #format (pnjang ls, jarak ls, panjang garis, jarak garis)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

#memberikan variasi simbol pada garis atau plot
#macam-maccam simbolnya https://matploblib.org/3.1.1/api/marker_api.html
ax.plot(x, x+9, color="blue", lw=3, ls='-', marker='+') #bisa kita singkat
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')  #lw = linewidth
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')  #ls = linestyle
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')
```
