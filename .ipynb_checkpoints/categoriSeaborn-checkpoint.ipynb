```python
import seaborn as sns
```

```python
#dummy data seaborn
tips = sns.load_dataset('tips')
tips.head()
```

```python
#memvisualisasi data mengunakan barplot
#dengan x adlah sex dan y adalah total_bill
dan datanya adalah tips
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
#memvisualisasikan data dengan count plot atau menghitung jumlah plot
#dengan x adalah sex dan data adalah tips
sns.countplot(x='sex',data=tips)
```

```python
#memvisualisasikan data dengan boxplot
#dimana x adalah day, y adlah total_bill dan data adalah variable tips
#palette adalah warna masing-masing plot
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
#memvisualisasikan data dengan boxplot secara orientasinya horizontal
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

```python
#memvisualisasikan data dengan boxplot
#hue akan menampilkan data dari warna data
sns.boxplot(x="day", y="total_bill", data=tips,palette='coolwarm',hue="smoker")
```

```python
#memvisualisasikan data dengan violin plot
sns.violinplot(x="day", y="total_bill", data=tips,palette='Set1',hue='sex')
```

```python
#split menggabungkan 2 tampilan yang sebelumnya terpisah
sns.violinplot(x="day", y="total_bill", data=tips,palette='Set1',hue='sex',split=True)
```

```python
#memvisualisasikan data dengan strip plot
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
#apabila jitter false akan membuat tampilanya mejadi satu baris
sns.stripplot(x="day", y="total_bill", data=tips, jitter=False)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True,hue='sex',palette='Set1')
```

```python
#dodge adalah split
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True,hue='sex',palette='Set1',dodge=True)
```

```python
#memvisualisasikan data dengan swarm plot
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
#memvisualisasikan data dengan violin plot
sns.swarmplot(x="day", y="total_bill", data=tips,palette='Set1',hue='sex',dodge=True)
```

```python
#kombinasi violin plot dengan swarmplot
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```

```python
#memvisualisasikan data dengan cAt plot
#kind adalah jenis tampilanya
sns.catplot(x="sex", y="total_bill", data=tips,kind="bar")
```
