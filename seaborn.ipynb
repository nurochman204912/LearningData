```python
import seaborn as sns
```

```python
#dummy data seaborn
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill']) #memvisualisasikan sebuah data ke bentuk histogram
```

```python
#kde False meghilangkan karnel atau garis pada plot
#bins jumlah batang yang ingin ditampilkan

sns.distplot(tips['total_bill'],kde=False,bins=10)
```

```python
#membuat diagram atau plot scatter
#dimana datanya adalah variable tips
#data x nya adalah total_bill 
data y nya adalah tip
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
#membuat diagram atau plot hex
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
#membuat diagram atau plot reg
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#menampilkan beberapa plot yang saling berhubungan
sns.pairplot(tips)
```

```python
#hue adalah pemberian warna yang berbedapada suatu data
#palette adalah warna dari hue
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
#hanya mengambar tanda hubunguntuk setiap titik pada distribusi
sns.rugplot(tips['total_bill'])
```
